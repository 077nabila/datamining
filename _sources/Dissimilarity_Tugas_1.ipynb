```python
# CONSTAN VARIABLE
DECREMENT_BY_ONE = 1
INCREMENT_BY_ONE = 1

CONTINGENCY_TABLE_VALUE = {
    "q" : (1,1),
    "r" : (1,0),
    "s" : (0,1),
    "t" : (0,0),
}
```

```python
def get_series(n, idx, series):
    return n.loc[(idx), series]
```

```python
def get_dissimilarity_dataset(n, series_index = [], series = []):
    first_series = get_series(n, series_index[0], series)
    second_series = get_series(n, series_index[1], series)
    dataset = pd.concat([first_series,second_series],axis=1)
    return dataset.T
```

```python
def count_contingency_value(n, start_index = 0, last_index = 1):

    CONTINGENCY_VALUE = {
        "q" : 0,
        "r" : 0,
        "s" : 0,
        "t" : 0,
    }

    column_range = n.shape[1]

    for column in range(column_range):
        for value in CONTINGENCY_TABLE_VALUE:
            item = list((tuple(n.loc[(start_index):(last_index), n.columns[column]]) == CONTINGENCY_TABLE_VALUE[value], value))
            if item[0] == True:
                if item[1] == "q":
                    CONTINGENCY_VALUE["q"] += 1
                if item[1] == "r":
                    CONTINGENCY_VALUE["r"] += 1
                if item[1] == "s":
                    CONTINGENCY_VALUE["s"] += 1
                if item[1] == "t":
                    CONTINGENCY_VALUE["t"] += 1

    return CONTINGENCY_VALUE
```

```python
# d(1,2)
d_1_2 = get_dissimilarity_dataset(n, [1,2], ["City", "Gender"])
```

```python
d_e_1_2 = count_contingency_value(d_1_2, 1, 2)
```
